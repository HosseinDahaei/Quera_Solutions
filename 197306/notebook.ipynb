```python
sources_counter = {}
for item in sources:
    sources_counter[item]=0
sources_counter
```

```python
for day in dirs:
    for src in os.listdir(f"./data/{day}"):
        temp = len(os.listdir(f"./data/{day}/{src}/"))
        sources_counter[src]+=temp
print(sources_counter)
```

```python
varzesh3 = []
for day in dirs:
    for src in os.listdir(f"./data/{day}"):
        if src != "varzesh3":
            continue
        temp = os.listdir(f"./data/{day}/{src}/")
        print(day,src,len(temp))
        for html_file in temp:
            varzesh3.append(f"./data/{day}/{src}/{html_file}")
len(varzesh3)
```

```python
from bs4 import BeautifulSoup 
targets = ["والیبال","کشتی","فوتبال"]
targets_counter = [0,0,0]
for file_name in varzesh3:
    # print(file_name)
    with open(file_name, 'r') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    p_tags = soup.find_all('p')

    for tag in p_tags:
        text = tag.text.replace("\""," ")
        # text = text.replace("\u200c"," ")
        text = text.replace("،"," ")
        text = text.replace(")"," ")
        text = text.replace("("," ")
        text = text.split(" ")
        for word in text:
            if targets[0] == word:
                targets_counter[0] += 1 
            elif targets[0] in word:
                print(targets[0]," in ",word)
            
            if targets[1] == word:
                targets_counter[1] += 1 
            elif targets[1] in word:
                print(targets[1]," in ",word)
            
            if targets[2] == word:
                targets_counter[2] += 1 
            elif targets[2] in word:
                print(targets[2]," in ",word)
            



print(targets_counter)
```

```python
special_day = []
for day in dirs:
    for src in os.listdir(f"./data/{day}"):
        if day != "2023-08-01":
            continue
        temp = os.listdir(f"./data/{day}/{src}/")
        # print(day,src,len(temp))
        for html_file in temp:
            special_day.append(f"./data/{day}/{src}/{html_file}")
len(special_day)
```

```python
import numpy as np

stop_words = list(np.load('./data/stopwords.npy'))
```

```python
from bs4 import BeautifulSoup 
big_text = ""
word_counter = {}
for file_name in special_day:
    # print(file_name)
    with open(file_name, 'r') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    h2_tags = soup.find_all('h2')
    for tags in h2_tags:
        text = tags.text
        big_text += " "+text+" "
    
len(big_text)
```

```python
words = sorted(big_text.split(" "))
len(words)
```

```python
index = 0
while index < len(words):
    word_counter[words[index]] = words.count(words[index])
    index += words.count(words[index])
print(word_counter)
```

```python
result = sorted(word_counter.items(),key=lambda a : a[1],reverse=True)
```
